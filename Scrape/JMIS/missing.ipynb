```python
import os

# Specify the path to the folder containing the PDF files
folder_path = r'JMIS'

# List all files in the specified folder
files = os.listdir(folder_path)

for filename in files:
    # Check if the file is a PDF
    if filename.endswith('.pdf'):
        # Normalize spaces: Replace two or more spaces with a single space
        new_filename = ' '.join(filename.split())
        
        # Remove space just before the ".pdf" if exists
        new_filename = new_filename.rstrip()  # Removes trailing whitespace
        if new_filename.endswith(' .pdf'):
            new_filename = new_filename[:-5] + '.pdf'  # Remove the last space before ".pdf"
        
        # Only proceed if the filename has changed
        if new_filename != filename:
            # Construct the full old and new file paths
            old_file_path = os.path.join(folder_path, filename)
            new_file_path = os.path.join(folder_path, new_filename)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f'Renamed "{filename}" to "{new_filename}"')
```

```python
import os
import pandas as pd

# Step 1: Read the CSV file to get the list of PDF names
csv_file_path = 'JMIS_data.csv'  # Update this to the path of your CSV file
pdf_names_df = pd.read_csv(csv_file_path)
pdf_names_list = pdf_names_df['Title'].tolist()

# Step 2: List all PDF files in the subfolder
pdf_folder_path = r'JMIS_PDF'  # Update this to your PDF subfolder path
pdf_files_in_folder = os.listdir(pdf_folder_path)
pdf_files_in_folder = [file for file in pdf_files_in_folder if file.endswith('.pdf')]

# Convert PDF names from the CSV file to the format they would have as filenames
pdf_names_in_csv_as_filenames = [((f"{name}.pdf").replace(":","").replace("?","").replace(",", "").replace("’",'').replace("…","").replace("/","").replace("“","").replace("”","")) for name in pdf_names_list]

# Step 3: Find which PDF names from the CSV are not in the subfolder
missing_pdfs = set(pdf_names_in_csv_as_filenames) - set(pdf_files_in_folder)

print(len(missing_pdfs))
print("PDF names from CSV not found in subfolder:", missing_pdfs)
```
